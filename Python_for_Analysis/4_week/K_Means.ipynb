```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

plt.style.use("ggplot")
plt.rcParams["figure.figsize"]= (16, 8)

%matplotlib inline
```

```python
X, y = make_blobs(n_samples=500, n_features=2, centers=10, random_state=42)
```

```python
def plot_scatter(labels, X, n):
    plt.scatter(X[labels == n][:, 0], X[labels == n][:, 1], label=str(n))
```

```python
for i in range(10):
    plot_scatter(y, X, i)

plt.legend()
```

```python
kmeans = KMeans(3)
```

```python
kmeans.fit(X)
```

```python
for i in range(kmeans.n_clusters):
    plot_scatter(kmeans.labels_, X, i)

plt.legend()
```

```python
clast = []

for i in range(2, 15):
    model = KMeans(i).fit(X)
    clast.append(model.inertia_)
```

```python
plt.plot(list(range(2, 15)), clast)
plt.xticks(list(range(2, 15)))
plt.legend()
```
